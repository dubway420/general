```python
# Import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def make_hardness_map(vmin, vmax):

    z = pd.read_excel('case7repeatfull.xlsx')

    x = np.linspace(0, 153, 152)
    y = np.linspace(0, 36, 37)

    xv, yv = np.meshgrid(x, y)

    # vmin = 200
    # vmax = 300

    n = plt.pcolormesh(xv, yv, z, cmap = plt.get_cmap('CMRmap'))
    n.set_clim(vmin, vmax)
    plt.axis('off')
    plt.colorbar(n)

    # Generate name based on vmin and vmax
    name = 'case7repeatfull_' + str(vmin) + '_' + str(vmax) + '.png'

    plt.imshow(z, cmap = plt.get_cmap('CMRmap'), vmin = vmin, vmax = vmax)
    plt.savefig(name, dpi = 300)

make_hardness_map(200, 300)
```
